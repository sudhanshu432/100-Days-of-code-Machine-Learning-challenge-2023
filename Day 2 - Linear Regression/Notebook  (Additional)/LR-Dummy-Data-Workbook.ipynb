```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("dummy_data.csv",delimiter=",")
```

```python
type(data)
```

```python
data.shape
```

```python
X = data[:,0].reshape(-1,1)
Y = data[:,1]
```

```python
X.shape
```

```python
from sklearn import model_selection
```

```python
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,Y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
alg1 = LinearRegression()
```

```python
alg1.fit(X_train,Y_train)
```

```python
m = alg1.coef_[0]
c = alg1.intercept_
```

```python
x = np.arange(30,75,0.1)
y = m*x+c
x_train_1d = X_train.reshape(-1)
x_test_1d = X_test.reshape(-1)
```

```python
plt.plot(x,y,color="red")
plt.scatter(x_train_1d,Y_train)
plt.grid()
plt.axis([20,80,20,130])
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Plot of fitted line with training data scattered")
plt.savefig("Dummy-data-training.png")
plt.show()
```

```python
plt.plot(x,y,color="red")
plt.scatter(x_test_1d,Y_test)
plt.grid()
plt.axis([20,80,20,130])
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Plot of fitted line with testing data scattered")
plt.savefig("Dummy-data-testing.png")
plt.show()
```
