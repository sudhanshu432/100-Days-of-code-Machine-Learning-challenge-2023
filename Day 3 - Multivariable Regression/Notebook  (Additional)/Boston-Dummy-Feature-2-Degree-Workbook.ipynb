```python
import pandas as pd
import numpy as np
from sklearn import datasets
```

```python
boston = datasets.load_boston()
```

```python
boston.DESCR.split("\n")
```

```python
X = pd.DataFrame(boston.data)
Y = pd.DataFrame(boston.target)
```

```python
X.columns = boston.feature_names
```

```python
X.columns
```

```python
X.describe()
```

```python
X2 = X.copy()
```

```python
for feature1 in X.columns:
    for feature2 in X.columns:
        comp_feature = feature1 + "_" + feature2
        X2[comp_feature] = X[feature1]*X[feature2]
```

```python
X2.describe()
```

```python
X2.columns
```

```python
X3 = X.copy()
for feature1 in X.columns:
    comp_feature = feature1 + "_" + feature1
    X3[comp_feature] = X[feature1]**2
```

```python
X3.columns
```

```python
from sklearn import model_selection
```

```python
X_train,X_test,Y_train,Y_test = model_selection.train_test_split(X,Y,random_state=0)
X2_train,X2_test,Y2_train,Y2_test = model_selection.train_test_split(X2,Y,random_state=0)
X3_train,X3_test,Y3_train,Y3_test = model_selection.train_test_split(X3,Y,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
alg1 = LinearRegression(normalize=True)
alg2 = LinearRegression(normalize=True)
alg3 = LinearRegression(normalize=True)
```

```python
alg1.fit(X_train,Y_train)
alg2.fit(X2_train,Y2_train)
alg3.fit(X3_train,Y3_train)
```

```python
print("Training score, for basic dataset:",alg1.score(X_train,Y_train))
print("Validation score for basic dataset:",alg1.score(X_test,Y_test))
print("Training score, for augmented dataset (all 2 degree):",alg2.score(X2_train,Y2_train))
print("Validation score for augmented dataset (all 2 degree):",alg2.score(X2_test,Y2_test))
print("Training score, for augmented dataset (homogenous 2 degree):",alg3.score(X3_train,Y3_train))
print("Validation score for augmented dataset (homogenous 2 degree):",alg3.score(X3_test,Y3_test))
```
